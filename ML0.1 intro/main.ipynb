```python
# 5

arr1 = np.random.randint(10, 30, (5, 5))
arr1[::2, ::2] = 5
arr1[1::2, 1::2] = 5
print(arr1, np.linalg.det(arr1), np.linalg.inv(arr1), "\n", sep="\n\n")

arr2 = np.empty((5, 5), dtype=int)
arr2.fill(5)
arr2 = np.triu(arr2)
print(arr2, np.linalg.det(arr2), np.linalg.inv(arr2), sep="\n\n")
```

```python
# 9

df = df_initial.drop(["Id", "DistrictId"], axis=1)
df = df.rename(
    columns={
        "Rooms": "комнаты",
        "Square": "площадь",
        "LifeSquare": "жилая площадь",
        "KitchenSquare": "площадь кухни",
        "Floor": "этаж",
        "HouseFloor": "этажей в доме",
        "HouseYear": "год постройки",
        "Ecology_1": "экология_1",
        "Ecology_2": "экология_2",
        "Ecology_3": "экология_3",
        "Social_1": "социальные_1",
        "Social_2": "социальные_2",
        "Social_3": "социальные_3",
        "Healthcare_1": "здравоохранение_1",
        "Helthcare_2": "здравоохранение_2",
        "Shops_1": "магазины_1",
        "Shops_2": "магазины_2",
        "Price": "цена",
    }
)
df
```

```python
# 10

print(df.iloc[:, -1])
print(df["цена"])
print(df.iloc[[0, 9, -2], :])
```

```python
# 11

tail = df.tail(10)
tail = tail.drop("цена", axis=1)
df = df.append(tail)
df["цена"] = df["цена"].fillna(df["цена"].mean())
```

```python
# 12

cols_tail = df.iloc[:, -5:]
cols_tail = cols_tail[cols_tail["цена"] >= cols_tail["цена"].mean()]
df = pd.merge(
    df, cols_tail, left_index=True, right_index=True, how="outer", suffixes=("", "_y")
)
df
```

```python
# 13

mean_price = df.groupby("этаж")["цена"].mean()
floors = df.groupby("этаж")["этаж"].count()

pd.DataFrame({"средняя цена": mean_price, "кол-во": floors})
```

```python
# 17

room_counts = df["Rooms"].value_counts()

other = sum([v for v in room_counts.values if v < 500])
room_counts = room_counts[room_counts >= 500]
room_counts["др."] = other

explode = [0.1 if i == room_counts.idxmax() else 0 for i in room_counts.index]

room_counts.plot.pie(explode=explode, autopct="%1.1f%%", startangle=90)

plt.title("Количество квартир с количеством комнат")
plt.show()
```
